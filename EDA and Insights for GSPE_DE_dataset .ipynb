# Exploratory Data Analysis (EDA) for GSPE_DE_Dataset

### Read the csv dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Load the data
df = pd.read_csv('GSPE_DE_dataset.csv')


#View the data
df.head()
```

### Basic information & Describe the data

```python
#Basic information

df.info()

#Describe the data

df.describe()
```

### Count of Duplicate Data

```python
df.duplicated().sum()
```

### Check the Duplicate values

```python
#Find the duplicates

df[df.duplicated('asst_id')].sum
```

### Delete the duplicate rows

```python
df1=df.drop_duplicates()
```

### New Dataset Information

```python
df1.duplicated().sum()
df1.info()
```

### Check for wrong spelled words

```python
unq_product_type=df1['product_type'].unique()

unq_contact_type=df1['contact_type'].unique()

unq_issue_type=df1['issue_type'].unique()

unq_topic_category=df1['topic_category'].unique()

unq_repair_type=df1['repair_type'].unique()

unq_diagnostics=df1['diagnostics'].unique()

unq_region=df1['region'].unique()

unq_country=df1['country'].unique()

unq_contact_type

## Voice and VOICE are same 
```

### Replace the wrong spelled words

```python
df1['contact_type'] = df1['contact_type'].replace(['Voice'], 'VOICE')
df1.head(20)
```

### Plot the graph for count of types of products used

```python

# count plot on single categorical variable
sns.countplot(x ='product_type', data = df1)

plt.title("Count of Product Type")
 
# Show the plot
plt.show()
```

### Plot the graph for medium contact

```python
sns.countplot(x ='contact_type', data = df1)
plt.title("Type of Contact")
plt.show()
```

### Plot the graph for customers based on region

```python
sns.countplot(x ='region', data = df1)
plt.title("Customers Based On Region")
plt.show()
```

### Plot the graph for types of repair

```python
sns.countplot(x ='repair_type', data = df1)
plt.title("Type of Repair")
plt.show()
```

### Check for Null Values

```python
df1.isnull().sum()
```

### Deal with Null Values

```python
df1["product_type"].fillna("Laptops", inplace = True)
df1["region"].fillna("Hogwarts", inplace = True)
df1["issue_type"].fillna("N/A", inplace = True)
df1["topic_category"].fillna("N/A", inplace = True)
df1["parts_sent"].fillna("N/A", inplace = True)
df1["repeat_parts_sent"].fillna("N/A", inplace = True)
df1["country"].fillna("N/A", inplace = True)
df1["contact_type"].fillna("N/A", inplace = True)
df1['agent_tenure_indays'].fillna(984.051823,inplace = True)
df1.isnull().sum()
```
